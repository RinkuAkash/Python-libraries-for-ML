```python
import numpy as np
from plotly.offline import iplot
import plotly.graph_objs as go
x=np.linspace(0,1,1000)
y=np.random.rand(1000)

trace= go.Scatter(
    x=x,
    y=y,
    mode='markers'
    )
data=[trace]
figure=go.Figure(data=data)
iplot(figure)
```
