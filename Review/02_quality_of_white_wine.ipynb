```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt


f,axs=plt.subplots(2,2,figsize=(9,8))
f.suptitle("Quality of White Wine",fontsize=14)

dataFrame=pd.read_csv("winequality-white.csv",sep=';')

sns.distplot(dataFrame['density'],kde=False,color='blue',ax=axs[0,0])

sns.distplot(dataFrame['residual sugar'],hist=False, rug=False,color='r',ax=axs[0,1])

sns.distplot(dataFrame['total sulfur dioxide'],hist=False, color='g',ax=axs[1,0])

sns.distplot(dataFrame['fixed acidity'],color='m',ax=axs[1,1])

plt.tight_layout()
```
