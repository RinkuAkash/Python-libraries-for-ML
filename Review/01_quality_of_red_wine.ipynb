```python
from matplotlib import pyplot as plt
import pandas as pd

dataFrame=pd.read_csv("winequality-red.csv",sep=';')

f,axs=plt.subplots(2,2,figsize=(10,7))


f.suptitle("Quality of Red Wine",color='maroon')

plt.subplot(221)
plt.scatter(dataFrame['density'],dataFrame['quality'],color='lightgreen',marker='.')
plt.xlabel("Density",color="blue")
plt.ylabel("Quality",color="green")
plt.grid(which='minor',color='skyblue')
plt.minorticks_on()

plt.subplot(222)
plt.bar(dataFrame['residual sugar'],dataFrame['quality'],color='gray')
plt.xlabel("Residual Sugar",color="blue")
plt.ylabel("Quality",color="green")

plt.subplot(223)
plt.scatter(dataFrame['fixed acidity'],dataFrame['quality'],color="red",marker='*')
plt.xlabel("Fixed Acidity",color='blue')
plt.ylabel("Quality",color='green')

plt.subplot(224)
plt.bar(dataFrame['total sulfur dioxide'],dataFrame['quality'],color='skyblue')
plt.xlabel("Total Sulfur Dioxide",color='blue')
plt.ylabel("Quality",color='green')

plt.show()
```
