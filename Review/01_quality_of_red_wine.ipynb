```python
from matplotlib import pyplot as plt
import pandas as pd

dataFrame=pd.read_csv("winequality-red.csv",sep=';')

dataFrame['density'].plot(kind='line', label="Density")
dataFrame['fixed acidity'].plot(kind='line',label="Fixed acidity")
dataFrame['total sulfur dioxide'].plot(kind='line',label="Sulfur dioxide")
dataFrame['residual sugar'].plot(kind='line',label="Residual sugar")

plt.grid(which='major',color='gray')
plt.grid(which='minor',color='skyblue')
plt.minorticks_on()

plt.xlabel("Data",color="blue")
plt.ylabel("Quality",color="green")
plt.title("Quality of Red Wine",color='maroon')
plt.legend()
plt.show()
```
