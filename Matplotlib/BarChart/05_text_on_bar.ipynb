```python
from matplotlib import pyplot as plt

languages=['Java','Python','PHP','JavaScript','C#','C++']
popularity=[22.2,17.6,8.8,8,7.7,6.7]

fig,axis=plt.subplots()
bars=axis.bar(languages,popularity,color='blue')
plt.xlabel("Programming Languages",color='maroon')
plt.ylabel("Popularity",color='maroon')
plt.title("Popularity of programming languages", color='black')
plt.minorticks_on()

for bar in bars:
    axis.text(bar.get_x()+bar.get_width()/2.,1*(bar.get_height()),'%f' % float(bar.get_height()), ha='center', va='bottom')
plt.show()
```
