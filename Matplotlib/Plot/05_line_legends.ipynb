```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to plot two or more lines on same plot with suitable legends of each line. 

'''

from matplotlib import pyplot as plt

plt.plot([1,2,3,4,5],[9,8,7,6,5],label="First Line")
plt.plot([1,2,3,4,5],[4,3,2,1,1],label="Second Line")
plt.plot([1,2,3,4,5],[0,0,9,0,0],label="Third Line")
plt.xlabel("y-axis")
plt.ylabel("x-axis")
plt.title("Display Legends")
plt.legend()
plt.show()
```
