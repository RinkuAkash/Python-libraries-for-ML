```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis and a title
'''

from matplotlib import pyplot as plt

x=list(map(int,input("Enter x-axis values : ").split()))
y=list(map(int,input("Enter y-axis values : ").split()))

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line (User axis)")
plt.show()
```
