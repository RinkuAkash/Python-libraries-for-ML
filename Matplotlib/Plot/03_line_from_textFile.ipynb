```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

'''

from matplotlib import pyplot as plt
import pandas as pd

df=pd.read_table("test.txt",delim_whitespace=True,names=('x','y'))
df.set_index('x',inplace=True)
plt.plot(df)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line (using text file)")
plt.show()
```
