```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to plot two or more lines with legends, different widths and colors. 

'''

from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")
x=[1,2,3,4,5]
plt.plot(x,[9,8,7,6,5],'r',label="First Line",linewidth=5)
plt.xticks(x)
plt.show()
style.use("grayscale")
plt.plot([1,2,3,4,5],[4,3,2,1,1],'c',label="Second Line",linewidth=8)
plt.show()
style.use("dark_background")
plt.plot([1,2,3,4,5],[0,0,9,0,0],'g',label="Third Line",linewidth=3)
plt.xlabel("y-axis")
plt.ylabel("x-axis")

plt.title("Display Legends")
plt.legend()
plt.show()
```
