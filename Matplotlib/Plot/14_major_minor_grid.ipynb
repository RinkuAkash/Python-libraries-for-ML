```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks. 

'''

from matplotlib import pyplot as plt
import pandas as pd

dataFrame=pd.read_csv("fdata.csv")
dataFrame.set_index('Date',inplace=True)
plt.plot(dataFrame,'g-o')
plt.grid(which='major', color='black',linestyle='-')
plt.grid(which='minor', color='gray', linestyle='--')
plt.xticks(dataFrame.index)
plt.minorticks_on()
plt.tick_params(which='both', top='off',left='off',right='off',bottom='off')
plt.title("Closing value of Alphabet lnc.")
plt.xlabel("Date")
plt.ylabel("Close")
plt.show()
```
