```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to create multiple plots. 
'''

from matplotlib import pyplot as plt
import numpy as np 

x=np.array([10,20,30,40,50])
y=np.array([23,35,48,14,41])

axis=plt.subplot2grid((8,8),(0,0),rowspan=4,colspan=4)
axis1=plt.subplot2grid((8,8),(4,0),rowspan=4,colspan=4)
axis2=plt.subplot2grid((8,8),(0,4),rowspan=4,colspan=4)
axis3=plt.subplot2grid((8,8),(4,4),rowspan=4,colspan=4)
axis.plot(x,y)
axis1.plot(x*2,y)
axis2.plot(x,y*2)
axis3.plot(x,y**3)
plt.show()
```
