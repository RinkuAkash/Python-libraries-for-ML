```python
'''
Created on 22/01/2020
@author: B Akash
'''
'''
problem statement:
Write a Python program to create multiple types of charts
'''

from matplotlib import pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.subplot(211)
plt.plot(x,y,'c')
plt.subplot(212)
plt.scatter(x,y**2,marker='o',color='g')
plt.show()
plt.subplot(221)
plt.bar(x,y,color='c')
plt.subplot(222)
plt.hist(x,bins=y)
plt.show()
plt.pie(x,labels=x,autopct='%1.2f%%')
plt.show()
```
